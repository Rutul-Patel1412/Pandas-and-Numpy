## JSON Data
Sometimes JSON data is complex and hard to understand. So, we can view data in online JSON viewer to understand it better.

```python
obj = """
{ "name":"Wes",
  "places_lived": ["United States", "Spain", "Germany"],
  "pet": null,
  "siblings": [{"name": "Scott", "age": 30, "pets":["zeus","zuko"]},
               {"name": "Katie", "age": 38,
                "pets": ["Sixes", "Cisco"]}]
}
"""
type(obj)
```

```python
# Use loads to convert JSON data from string to dictionary
import json
result = json.loads(obj)
type(result)
```

```python
# Use dumps to convert JSON data from dictionary to string
asjson =  json.dumps(result)
type(asjson)
```

```python
result['siblings']
# It has two dictionaries
```

```python
# In siblings we are interested in only age and pets column
siblings = pd.DataFrame(result['siblings'], columns=['age','pets'])
siblings
```

## Web APIs  (requires internet)
In simple terms, API is a connectivity between two homogeneous or heterogeneous(like one in python and other in java) systems which will take request and will give you response. We can not build any application without API as every system will be different system, so we have to establish channel or communication between them using API.

```python
import requests
url = 'https://api.github.com/repos/pandas-dev/pandas/issues'
resp = requests.get(url)
resp
```

```python
data = resp.json()
data[2]['user']
```

```python
# Use online jsonviewer
# Get all data in 'user' 'id'
l = []
for i in range(len(data)):
    l.append(data[i]['user']['id'])
l
```
