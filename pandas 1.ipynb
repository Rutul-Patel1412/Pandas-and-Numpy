## Get Tabular data from a website
Apde je website mathi data levo che, ae website ma table sivay bi ghani badhi vastuo hoy like ads, description, other informations, etc. But we don't want other thing, we only want the table from that particular website. So, our machine extract data which is in table tag(from html) only. And in this way we get table from that website and avoid other data. 
Note : Using PANDAS we can only extract tabular format data. For other kinds of data we need bs4 (beautiful soup) library. 

```python
# We can directly give url in bracket also.
url = "Website path"
BB_data = pd.read_html(url)

# Notation is given in list format by default.....because ek website ma ek thi vadhare table pan hoi sake and ae badha list format ma store karine ape. 
# [dataframe1,dataframe2,....]
# Try type(BB_data) 
BB_data
```

```python
# To get a particular dataframe from list we need to give index in []
BB_data[0]
# Or BB_data[1] and so on
```

```python
# Sudhanshu ae je data frame extract kari ae 675 rows * 30 columns ni che. Akha 30 columns apda screen ma na avi shake. So, apde koi particular columns jova hoy to below commands try karva
df = BB_data[0]

# To see a single column. NOTE:  For particular column, after extracting, if we check type(df['Column name']) it will give type as "SERIES"
df['Particular Column name']
## NOTE : If we give the column name in list and then check type, then it will give type as dataframe
df[['Particular Column name']]

# To see a particular two columns
## This will give key error 
df['Column name 1','Column name 2']
## We need to pass above column names as list and then it will work fine
df[['Column name 1','Column name 2']]

# To see list of columns
df.columns
```
